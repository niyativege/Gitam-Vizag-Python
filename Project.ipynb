## 1. Display the count of matches based on the year wise

```python
import pandas as pd 
filePath='C:/Users/Admin/Desktop/problem solving and programming/DataFiles/Match.csv'
def readCsvData(filePath):
    return pd.read_csv(filePath)
df = readCsvData(filePath)
print(df)
```

```python
pd.DataFrame(df.groupby('Season_Year')['Match_SK'].count())
```

## 2.Plot a bar graph  no. of matches won by each team in the year  2011


```python
import pandas as pd
import matplotlib.pyplot as plt
pt=df[df['Season_Year']==2011]
lr=pt['match_winner'].value_counts()
pd.DataFrame(lr)
print(lr)
print()
teams=[]
teams=lr.index.tolist()
plt.barh(teams,lr,color="Red")
```

## 3.Print  ManOfMatch count of each player in  Hyderabad stadium

```python
df2=df.copy()
new_df=pd.DataFrame(df2[df2['City_Name'].isin(['Hyderabad'])])
pd.DataFrame(new_df.groupby('ManOfMach')['Match_SK'].count())
```
